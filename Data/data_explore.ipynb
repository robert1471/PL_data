
Let's explore FPL data generated by data_obtain and data_set_up

```python
import pandas as pd
```

First, load all dataframes into a dictionary as:

```python
gameweek_dfs = {}
for i in range(1, 48):
    gameweek_dfs["gw_{}".format(i)] = pd.read_csv("../Data/gw_{}.csv".format(i))
```

Things can then be obtained from here in loops like players total data. Here is an example for de bruyne.

```python
de_bruyne = pd.DataFrame()
for i, gameweek in enumerate(gameweek_dfs):
    week_data = gameweek_dfs[gameweek].loc[gameweek_dfs[gameweek]['ids'] == 215]
    de_bruyne  = de_bruyne.append(week_data)
    
  
```

But probably best to define a nice function to do this for any column

```python
def cat_to_df(gw_dict, category):
    cat_df = pd.DataFrame()
    for i, gameweek in enumerate(gw_dict):
        # define a working df
        df = gameweek_dfs[gameweek][['ids', category]]
        df.rename(columns={category: gameweek}, inplace=True)
        if i == 0:
            # add ids and goals scored from requested data to df
            cat_df = pd.concat([cat_df, df], axis=1)

        else:
            # append only goals scored
            cat_df = pd.concat([cat_df, df[gameweek]], axis=1)
    return cat_df
```

This will then let us mess about with dataframes of different categories such as goals conceded

```python
goals_conceded = cat_to_df(gw_dict=gameweek_dfs, category='goals_conceded')

max_conceded = goals_conceded.max()
mean_conceded = goals_conceded.mean()
gw_9_summary = goals_conceded["gw_9"].describe()


print(gw_9_summary)
```
