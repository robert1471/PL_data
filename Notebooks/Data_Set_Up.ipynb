First the imports

```python
import requests
import pandas as pd
import numpy as np
```

Want to add each team name to df 

```python
url_name = 'https://fantasy.premierleague.com/api/bootstrap-static/'

r = requests.get(url_name)
x = r.headers
json = r.json()
json.keys()
```

Get players and teams in coded dicts

```python
teams = {}
players = {}
player_team_dict = {}
# Teams
for team in json['teams']:
    teams[team['id']] = team['name']
    

for player in json['elements']:
    players[player['id']] = player['web_name']
    player_team_dict[player['web_name']] = teams[player['team']]
    print(player['web_name'], player['team'], teams[player['team']])
```

Then define a function that adds the player names to each csv

```python
def add_player_team_names(file, players=None, player_team=None):
    
    gw_df = pd.read_csv(file)
    for i, player in gw_df.iterrows():
        gw_df.loc[i, 'player_name'] = players[player['ids']]
        gw_df.loc[i, 'team_name'] = players[player['ids']]
        
    for i, team in gw_df.iterrows():
        gw_df.loc[i, 'team_name'] = player_team[gw_df.loc[i, 'player_name']]


    cols = gw_df.columns.tolist()
    cols = cols[1:]
    cols = cols[-2:] + cols[:-2]
    
    gw_df = gw_df[cols]

    gw_df.to_csv(file)
```

```python
# For all gameweeks
for gw in range(1, 48):
    add_player_team_names("../Data/gw_{}.csv".format(gw), players=players, player_team=player_team_dict)
```
